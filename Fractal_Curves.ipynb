```python
################ WELCOME TO FRACTAL CURVES! ###################

import math                       # 파이썬의 기본 수학 연산 패키지
from  numpy import *              # 벡터 및 행렬 연산을 위한 패키지
import matplotlib.pyplot as plt   # 그래프 및 시각화를 위한 패키지
```

```python
############# 대표적인 프랙탈 곡선인 코흐 곡선을 그려봅시다! ###############

def draw_fractal_curve(v,n):
  ppo = array([0,1])
  for i in range(0,n):
    npo = array([])
    for k in range(0,len(ppo)-1):
      npo = append(npo,(ppo[k+1]-ppo[k])*v+ppo[k])
    npo = append(npo,1)
    ppo=npo
  fig, axis = plt.subplots()
  axis.plot( real(ppo), imag(ppo), lw = 1, c = 'k')
  axis.set_aspect('equal')
  fig.suptitle("Your Fractal Curve!", fontsize=20)
  fig.set_size_inches(21,7)
  fig.tight_layout()


####### 원하는 대로 세팅하신 후 Ctrl+Enter! ######

v1 = array([                                     # 프랙탈 생성 법칙
          0, 
          1/3*exp(0*1j), 
          2/3*math.cos(pi/6)*exp(pi/6*1j), 
          2/3*exp(0*1j)
          ])
v2 = array([
          0, 
          3/10*exp(0*1j), 
          4/10*exp(pi/8*1j), 
          5/10*exp(-pi/7*1j), 
          8/10*exp(0*1j)
          ])
v3 = array([
          0, 
          6/20*exp(pi/24*1j),
          7/20*exp(3*pi/24*1j), 
          10/20*exp(-2*pi/24*1j), 
          12/20*exp(-3*pi/24*1j), 
          13/20*exp(2*pi/24*1j), 
          17/20*exp(1*pi/24*1j)
          ])  

N = 6                                           # 프랙탈 생성을 반복할 횟수 (6을 넘지 않는 것을 추천합니다!!!)

###############################################

draw_fractal_curve(v1,N)      # 프랙탈 곡선 그리기!
```
